### 2.3 - Advanced: Raise rate limit on Github <a class="anchor" id="bullet2x3"></a>
##### [back to TOC](#TOC)

In this Text-Fabric dataset, we included a 'complete.zip' file for each release, which allows the Text-Fabric code to download all data together. This function is called 'express load' but can only be used for the latest release. When using other releases, or when using a Text-Fabric dataset that does not offer a 'complete.zip', users may hit the download rate limits of GitHub, especially when loading Text-Fabric datasets that consist of many files. In such cases, it may be necessary to increase the rate limit for GitHub. To do so, you need a personal GitHub account which allows for the creation of a `GHPERS`, a personal GitHub token. [See instructions](https://annotation.github.io/text-fabric/tf/advanced/repo.html#increase-the-rate-limit) on acquiring and setting the `GHPERS` variable. See [here](https://www.howtogeek.com/789660/how-to-use-windows-cmd-environment-variables/#autotoc_anchor_2) if you want to set the variable on Windows using the command prompt.
