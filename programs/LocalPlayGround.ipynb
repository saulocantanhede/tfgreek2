```python
from tf.app import use
```

```python
# load the N1904 app and data from local disc
PLAY = use("app:/TF/app", hoist=globals())
```

```python
# Set default view in a way to limit noise as much as possible.
PLAY.displaySetup(condensed=True, queryFeatures=False,multiFeatures=False)
```

```python
SentenceQuery = '''
verse book=John chapter=1 verse=1
'''

SentenceResults = PLAY.search(SentenceQuery)
```

```python
HideNonWG={"clause","phrase","subphrase","group"}
MorphFeatures={'mood','gender','case','number','tense'}
PLAY.show(SentenceResults, start=1, end=1, hiddenTypes=HideNonWG, extraFeatures=MorphFeatures)
```

```python
HideWG={"wg","subphrase"}
PLAY.show(SentenceResults, start=1, end=1, hiddenTypes=HideWG, extraFeatures=MorphFeatures)
```
